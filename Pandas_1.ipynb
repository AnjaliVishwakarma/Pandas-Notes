# Operations on Dataset

```python
df['temperature'].max()  # maximum temp
```

```python
df['temperature'].min()  # min temp
```

```python
df['temperature'].mean() # avg temp
```

```python
df['temperature'].std()  # standard deviation
```

```python
df.describe()   # it print the statistics data present in our dataset and gives perencentiles data also
```

```python
df[df.temperature>=32] # it print the temp value which is greater than 32
```

```python
df[df.temperature == df.temperature.max() ]  # it gives the max temp value it just like sql query condition 
```

```python
# Or
df[df.temperature == df['temperature'].max()]
```

```python
# if we didn't want to print entire row or we wanna print sinlge value than we do that
df['day'][df.temperature == df['temperature'].max()]
```

```python
df[['day','temperature']][df.temperature == df['temperature'].max()]
```
