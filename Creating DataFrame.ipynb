# Different Ways of Creating DataFrame,
* Using CSV
* Using Excel
* From python dictionary
* From list of tuples
* From list of dictionary


```python
import pandas as pd
```

```python
#CSV

df= pd.read_csv("weather1.csv")
df
```

```python
# Excel

df = pd.read_excel("weather1.csv.xlsx","Sheet1")
df
```

```python
# Python Dictionary

data_res = {
    "day": ['1/1/2017','1/4/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    "temperature": [32,35,28,24,32,31],
    "windspeed": [6,7,2,7,4,2],
    "event":['Rain','Sunny','Snow','Snow','Rain','Sunny']
    
}

df = pd.DataFrame(data_res)
df
```

```python
# List of touples
Data = [
    
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',26,2,'Snow')
    
]

df = pd.DataFrame(Data, columns = ['Day','Temperature','WindSpeed','Event'])
df
```

```python
Data = [
    
    {'Day': '1/1/2017', 'Temperature': 32, 'Windspeed': 6, 'Event': 'Rain'},
    {'Day': '1/2/2017', 'Temperature': 35, 'Windspeed': 7, 'Event': 'Sunny'},
    {'Day': '1/3/2017', 'Temperature': 26, 'Windspeed': 2, 'Event': 'Snow'}
    
]
df = pd.DataFrame(Data)
df
```
